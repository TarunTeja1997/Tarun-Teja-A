```python
def draw_boxplot(df, X, xlab, title):

    x       =   df[X]
    AHD     =   df['ProductCategory']
    
    print("\n Summary statistics for  " + xlab + "\n")
    print(df[X].groupby(df['ProductCategory']).describe().T)

    sns.set_style("whitegrid")
    ax = sns.boxplot(x = df[X] ,y = df['ProductCategory'])
    ax.set_title(title)
    plt.show()  
```
